```python
import requests

s = requests.Session()

# Посетите сайт
s.get('http://opp-api.ozon.ru')

# Сохраните cookies в файл
with open('cookies.txt', 'w') as f:
    for cookie in s.cookies:
        f.write(f'{cookie.name}={cookie.value}\n')
```

```python
cookies = {}
with open('cookies.txt', 'r') as f:
    for line in f:
        if "=" in line:
            name, value = line.strip().split('=', 1)
            cookies[name] = value
for c, v in cookies.items():
    print(c,v)
```

```python
import json
from seleniumwire import webdriver
from selenium.webdriver.chrome.options import Options
import time


chrome_options = Options()
chrome_options.add_argument('--disable-blink-features=AutomationControlled')

browser_driver = webdriver.Chrome(options=chrome_options)
browser_driver.get('https://market.yandex.ru')

time.sleep(5)

cookies = []
for cookie in browser_driver.get_cookies():
    cookies.append(cookie)

with open('cookies.json', 'w') as f:
    json.dump(cookies, f)

time.sleep(15)

browser_driver.quit()
```
