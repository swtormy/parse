```python
# получаение квадратов покрывающих территорию России
def get_square_array(square_size: int = 300) -> dict:
    south_lat = 41.185
    north_lat = 81.857
    west_lon = 19.641
    east_lon = 190.95
    num_squares_lat = int(geodesic((south_lat, west_lon), (north_lat, west_lon)).km // square_size)
    num_squares_lon = int(geodesic((south_lat, west_lon), (south_lat, east_lon)).km // square_size)
    lats = np.linspace(south_lat, north_lat, num_squares_lat)
    lons = np.linspace(west_lon, east_lon, num_squares_lon)
    squares = []
    for i in range(num_squares_lat-1):
        for j in range(num_squares_lon-1):
            sw = (lats[i], lons[j])
            ne = (lats[i+1], lons[j+1])
            square_str = {
                'ne': f'{ne[0]:.5f},{ne[1]:.5f}',
                'sw': f'{sw[0]:.5f},{sw[1]:.5f}'
            }
            squares.append(square_str)
    return {'square_size': square_size, 'squares': squares}

# определение квадратов с объявлениями
def squared_filtering(squares: list, url: str, params: dict, cookies: dict, headers: dict) -> list:
    new_params = params
    rigth_coordinates = []
    count_squares = len(squares)
    for index, sq in enumerate(squares):
        try:
            new_params.update({
                'ne': sq['ne'],
                'sw': sq['sw']
            })
            response = requests.get(url, params=params, cookies=cookies, headers=headers)
            pagination = response.json()['pagination']
            total = pagination['total']
            sq_obj = {}
            if total > 0:
                sq_obj.update({
                    'coordinates': sq,
                    'pagination': pagination
                })
                rigth_coordinates.append(sq_obj)
                message = f'Квадрат {index + 1} из {count_squares}. Найдено {total} объявлений.'
                print(message.ljust(150), end='\r')
            message = f'Квадрат {index + 1} из {count_squares}. Объявлений нет.'
            print(message.ljust(150), end='\r')
        except KeyError:
            errors = ';'.join([','.join(error['errors']) for error in response.json()['errors']])
            print(errors.ljust(150), end='\r')
        except Exception as err:
            print('\n',response.status_code, err)
            break
    return rigth_coordinates

# пасринг простого квадрата
def simple_square_parsing(squares: list, url: str, params: dict, cookies: dict, headers: dict, file_path_prefix: str = f"./data/data_{int(time.time())}"):
    result = []
    for index, sqr in enumerate(squares):
        items_array = []
        error_messages = []
        response_errors = []
        status = 200

        offset = 0
        limit = 30
        coordinates = sqr['coordinates']
        pagination = sqr['pagination']
        total = pagination['total']
        message = f"Квадрат {sqr['coordinates']} {index+1} из {len(squares)}:"
        print(message.ljust(150), end='\r')

        while offset < total:
            message = f'Квадрат {index+1} из {len(squares)}: Выполняю работу в оффсете {offset} из {total}. Ошибки: {len(response_errors)}. Статус последней ошибки: {status}'
            print(message.ljust(150), end='\r')
            try:
                params['ne'] = coordinates['ne']
                params['sw'] = coordinates['sw']
                params['offset'] = offset
                params['limit'] = limit
                response = requests.get(url, params=params, cookies=cookies, headers=headers)
                status = response.status_code

                if status == 200:
                    response_errors = []
                    for item in response.json()['result']['items']:
                        items_array.append(item)
                    offset += limit
                else:
                    response_errors.append(f"Статус запроса {response.status_code} в оффсете {offset}")
                    error_messages.append(f"Статус запроса {response.status_code} в оффсете {offset}")

                time.sleep(random.uniform(1, 3))
            except Exception as e:
                response_errors.append(f"Ошибка: {str(e)} в оффсете {offset}")
                error_messages.append(f"Ошибка: {str(e)} в оффсете {offset}")
                offset += limit

        result_item = {
            'coordinates': coordinates,
            'pagination': pagination,
            'items_array': items_array
        }
        result.append(result_item)

        file_path = f"./data/data_{int(time.time())}_{index}.json"  
        with open(file_path, 'w') as file:
            json.dump(result_item, file, ensure_ascii=False)
            file.write('\n') 

    return result

# пасринг сложного квадрата
def split_and_parse(squares, max_total=2000):
    for index, sqr in enumerate(squares):
        if sqr['pagination']['total'] > max_total:
            smaller_squares = split_square(sqr['coordinates'], 1) 
            smaller_squares = squared_filtering(
                squares=smaller_squares, 
                url=base_url,
                params=config.params,
                cookies=config.cookies,
                headers=config.headers
            )  
            split_and_parse(smaller_squares)
        else:
            message = f'Квадрат {index+1} из {len(squares)}'
            print(message.ljust(150), end='\r')
            simple_square_parsing([sqr], url=base_url, params=config.params, cookies=config.cookies, headers=config.headers, file_path_prefix=f"./data/data_d_{int(time.time())}") 

# разделение квадрата
def split_square(square, levels):
    if levels == 0:
        return [square]
    
    ne_lat, ne_lon = map(float, square['ne'].split(','))
    sw_lat, sw_lon = map(float, square['sw'].split(','))
    
    mid_lat = (ne_lat + sw_lat) / 2
    mid_lon = (ne_lon + sw_lon) / 2
    
    square1 = {'ne': f'{mid_lat:.5f},{mid_lon:.5f}', 'sw': f'{sw_lat:.5f},{sw_lon:.5f}'}
    square2 = {'ne': f'{ne_lat:.5f},{mid_lon:.5f}', 'sw': f'{mid_lat:.5f},{sw_lon:.5f}'}
    square3 = {'ne': f'{mid_lat:.5f},{ne_lon:.5f}', 'sw': f'{sw_lat:.5f},{mid_lon:.5f}'}
    square4 = {'ne': f'{ne_lat:.5f},{ne_lon:.5f}', 'sw': f'{mid_lat:.5f},{mid_lon:.5f}'}
    
    squares = split_square(square1, levels-1)
    squares.extend(split_square(square2, levels-1))
    squares.extend(split_square(square3, levels-1))
    squares.extend(split_square(square4, levels-1))
    
    return squares
```
